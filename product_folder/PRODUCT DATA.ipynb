Create a DataFrame from the list of dictionaries below:

```python
#let display our DataFrame by importing Pandas First

import pandas as pd

list_of_products=[{'product_id':23, 'name':'computer', 'wholesale_price': 500, 'retail_price':1000, 'sales':100}, 
{'product_id':96, 'name':'Python Workout', 'wholesale_price': 35,'retail_price':75, 'sales':1000},
{'product_id':97, 'name':'Pandas Workout', 'wholesale_price': 35, 'retail_price':75, 'sales':500},
{'product_id':15, 'name':'banana', 'wholesale_price': 0.5,'retail_price':1, 'sales':200},
{'product_id':87, 'name':'sandwich', 'wholesale_price': 3,'retail_price':5, 'sales':300}]

df= pd.DataFrame.from_records(list_of_products,index='product_id')
df
```

```python
#Calculate the Total Profit for each product using the formula 

df['net_revenue_product']=(df['retail_price']-df['wholesale_price'])*df['sales']
df
```

```python
#How much total net revenue you received from all of these sales?

df['Total_revenue']=df['net_revenue_product'].sum()
df
```

```python
#What product is product retail price more than twice the wholesale price?

df['Product_more_than_twice']=df[df['retail_price']>2*df['wholesale_price']]['name']
df
```

```python
#How much did the store make from food vs. computers vs. books?
#-Food:
df['Food_net_revenue']=df[df['name'].isin(['banana','sandwich'])]['net_revenue_product'].sum()
df
```

```python
#-Computer:

df['Computer_net_revenue']=df[df['name']=='computer']['net_revenue_product'].sum()
df
```

```python
#-Books:
df['Book_net_revenue']=df[df['name'].isin(['Python Workout','Pandas Workout'])]['net_revenue_product'].sum()
df
```

```python
#Calculate the new net revenue after 30% discount on wholesales price

df['new_wholesales_price']=df['wholesale_price']*(1-0.3)
df['new_net_revenue_per_product']=(df['retail_price']-df['new_wholesales_price'])*df['sales']
df
```
